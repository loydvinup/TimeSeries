```python
out_col=['age','duration','campaign','previous']
```

```python
for i in out_col:
    Q1=df[i].quantile(0.25)
    Q3=df[i].quantile(0.75)

    IQR=Q3-Q1
    lower=Q1-1.5*IQR
    upper=Q3+1.5*IQR

    df=df[(df[i]>=lower) & (df[i]<=upper)]
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate x values for the curve
x = np.linspace(-7, 7, 200)  # You can adjust the range and number of points as needed

# Compute the sigmoid values for the x values
y = sigmoid(x)

# Plot the sigmoid curve
plt.figure(figsize=(8, 6))
plt.plot(x, y, color='blue', label='Sigmoid Curve')
plt.title('Sigmoid Curve')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.axhline(0.5, color='red', linestyle='--', label='y = 0.5', linewidth=1)  # Add a horizontal line at y=0.5 for reference
plt.legend()
plt.grid(True)
plt.show()
```
